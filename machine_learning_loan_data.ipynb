```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('loan_data.csv')
df
```

```python
# informatii despre coloane
df.info()
```

```python
#reprezentam sexul cu valorile 0 si 1
df['Gender'] = df['Gender'].map({'Male' : 0,'Female' : 1})
```

```python
df
```

```python
#verificam valorile la Married
df['Married'].value_counts()
```

```python
#reprezentam married cu valorile 0 si 1
df['Married'] = df['Married'].map({'Yes' : 0,'No' : 1})
```

```python
df
```

```python
#verificam education
df['Education'].value_counts()
```

```python
#reprezentam education cu valorile 0 si 1
df['Education'] = df['Education'].map({'Graduate' : 0,'Not Graduate' : 1})
```

```python
df
```

```python
#verificam valorile self_employed
df['Self_Employed'].value_counts()
```

```python
#reprezentam Self_Employed cu valorile 0 si 1
df['Self_Employed'] = df['Self_Employed'].map({'No' : 0,'Yes' : 1})
```

```python
df
```

```python
#verificam valorile Property_area
df['Property_Area'].value_counts()
```

```python
# coloanele care au mai multe valori le inlocuim cu dummies variable
df = pd.get_dummies(df, columns=['Loan_ID','Property_Area',], drop_first=True)
```

```python
df
```

```python
# importam bibliotecele
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, plot_confusion_matrix
```

```python
# vedem rezultatel dupa Loan_Status
X = df.drop(['Loan_Status'], axis=1).values
y = df['Loan_Status'].values
```

```python
# antrenam
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# predictia la decision
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
# acurateteaa la testare
y_pred = tree.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
# acuratetea la antrenare
y_pred_train = tree.predict(X_train)
accuracy_score(y_pred_train, y_train)
```

```python
# graficul pentru tree
plot_confusion_matrix(tree, X_test, y_test)
```

```python
# predictia la forest
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
# acuratetea la testare
y_pred = forest.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
# acuratetea la antrenare
y_pred_train = forest.predict(X_train)
accuracy_score(y_pred_train,y_train)
```

```python
# graficul pentru forest
plot_confusion_matrix(forest, X_test, y_test)
```

```python
# predictia la logit
logit = LogisticRegression(max_iter=10000)
logit.fit(X_train,y_train)
```

```python
# acuratetea la testare
y_pred = logit.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
# acuratetea la antrenare
y_pred_train = logit.predict(X_train)
accuracy_score(y_train,y_pred_train)
```

```python
# graficul pentru logit
plot_confusion_matrix(logit, X_test, y_test)
```

```python
# predictia la knn
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# acuratetea la testare
y_pred = knn.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
# acuratetea la antrenare
y_pred_train = knn.predict(X_train)
accuracy_score(y_pred_train, y_train)
```

```python
# graficul pentru knn
plot_confusion_matrix(knn, X_test, y_test)
```

```python
#predictia la gausian
gauss = GaussianNB()
gauss.fit(X_train, y_train)
```

```python
# acuratetea la testare
y_pred = gauss.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
#acuratetea la antrenare
y_pred_train = gauss.predict(X_train)
accuracy_score(y_pred_train, y_train)
```

```python
# graficul la gauss
plot_confusion_matrix(gauss, X_test, y_test)
```

```python
# definim o lista cu toate modelele
# afisare acuratete la testare si antrenare
def fit_all(X, y):
    models = [LogisticRegression(max_iter=1000), KNeighborsClassifier(),
              DecisionTreeClassifier(), RandomForestClassifier(),
              GaussianNB()]
    names = ['Logit', 'KNN', 'CART', 'FOREST', 'GAUSS']
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    for i in range(len(models)):
        print(names[i])
        models[i].fit(X_train, y_train)
        y_pred = models[i].predict(X_test)
        y_pred_train = models[i].predict(X_train)
        train_acc = accuracy_score(y_train, y_pred_train)
        test_acc = accuracy_score(y_test, y_pred)
        print(f"train acc - {train_acc}")
        print(f"test acc - {test_acc}")
```

```python
fit_all(X,y)
```
