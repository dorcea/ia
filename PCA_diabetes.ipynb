```python
#PCA - reducere dimensionala
# vizualizarea datelor, tabele mari cu date
# vizualizarea in 2d 3d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import seaborn as sns
```

```python
# creem o matrice cu 500 rinduri si 3 coloane
# cu valori intre 0 si 1
X = np.random.uniform(size=(500, 3))
```

```python
#afisam
X
```

```python
# vizualizare cod in 3d sa vedem punctele cum sunt aranjate in spatiu
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1],X[:, 2])

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
#reducerea pina la 2 coloane---de exemplu 2
pca = PCA(n_components = 2)
```

```python
pca.fit(X)
```

```python
pca_X = pca.transform(X)
```

```python
pca_X
```

```python
# reducem matricea din 3d in 2d
#111 inseamna cu 3 dimensini
#np.zeros(len(pca_X) am facut matricea doar cu 0
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_X[:, 0], pca_X[:, 1], np.zeros(len(pca_X)))
plt.legend()
plt.show()
```

```python
# matrice mai mica dar este mai compacta
plt.scatter(pca_X[:, 0], pca_X[:, 1])
```

```python
# luam ca exemplu data setul cu heart
df =pd.read_csv('diabetes.csv')
df
```

```python
# aflam nr optimal de valori
X = df.drop(['Outcome'], axis=1).values
```

```python
X
```

```python
pca = PCA()
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
# sa pastram cit mai multa informatie aproape 99%
# 14 inseamna valorile de la array...de numarat
cumsum = np.cumsum(pca.explained_variance_ratio_)
xi = range(1,9)
plt.plot(xi, cumsum)
# linia ne spune avind atitea componente, 2 --0,75, 3 -0,90
# acele 3 curbe
```

```python
pca = PCA(n_components=5)
pca.fit(X)
pca_X = pca.transform(X)
```

```python
# vizualizarea datelor, tabele mari cu date
y = df['Outcome'].values
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_X[:, 0], pca_X[:, 1], pca_X[:, 2], c=y)
plt.legend()
plt.show()
```
